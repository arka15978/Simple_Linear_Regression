```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/1.01. Simple linear regression.csv")
a = df['X']
b = df['Y']
A = 0
B = 0
for i in range(10000):
  sum1 = 0
  sum2 = 0
  for i in range(len(a)):
    sum1 += 2*(A*a[i]/1000 + B - b[i]/1000)*a[i]/(len(a)*1000)
    sum2 += 2*(A*a[i]/1000 + B - b[i]/1000)/len(a)
  A = A - 0.0001*sum1
  B = B - 0.0001*sum2
print(A,B)


 
 
 


```
